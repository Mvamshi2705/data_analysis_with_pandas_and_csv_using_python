```python
import pandas as pd
import folium
```

```python
df=pd.read_csv("india_historical_places.csv")
```

```python
print("==== HISTORICAL PLACES DATA ====")
print(df)
```

```python
print("\nTotal Historical Places:", len(df))
```

```python
print("State Wise Count:")
print( df["State"].value_counts())
```

```python
print("Oldest Historical Place:")
print(df.loc[df["YearBuilt"].idxmin()])
```

```python
print("Mughal Empire Places:")
print(df[df["Dynasty"]=="Mughal Empire"])
```

```python
#----------------------------------------
# CREATE INDIA MAP
#----------------------------------------
```

```python
india_map=folium.Map(location=[22.9734, 78.6569], zoom_start=5)

for index, row in df.iterrows():
    popup_text =f"""
<b>{row['Place']}</b><br>
{row['Description']}<br>
State: {row['State']}<br>
Building: {row['YearBuilt']}<br>
Dynasty: {row['Dynasty']}
"""
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_text,
    ).add_to(india_map)

india_map.save("india_historical_map.html")
print("India historical map created successfully")
```
